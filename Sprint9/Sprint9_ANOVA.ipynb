```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

aids4 = 'aids4.csv'

df = pd.read_csv(aids4, header=1)
df.head()
```

```python
data = {
    'achieved_error': [1-0.817, 1-0.866, 1-0.592, 1-0.876, 1-0.261, 0, round(0.4465/47.028, 3), round(1844.993/3938246.456, 3), round(33970.6133/206855.817, 3), round(33144.565/206855.817, 3), round(33661.3486/206855.817, 3)],
    'num_pit_samples': [150, 1000, 250, 1000, 150, 50, 250, 250, 100, 500, 150],
    'take_num_samples_when_full': [0, 0, 0, 1, 0, 0, 0, 0, 0, 1, 0],
    'normal_backtransform': [False, False, True, True, True, True, False, True, True, True, False],
    'trainable_bandwidths': [True, False, False, True, True, True, True, True, False, False, True],
    'task_type': ['classification', 'classification', 'classification', 'classification', 'classification', 'regression', 'regression', 'regression', 'Regression', 'regression',  'regression'],
    'num_classes': [2, 2, 2, 2, 114, 0, 0, 0, 0, 0, 0],
    'num_features': [22, 22, 95, 10, 15, 59, 385, 248, 8, 8, 8],
    'instances': [2139, 2139, 6819, 19020, 114000, 39644, 53500, 36043, 20640, 20640, 20640],
    'pit_ordinary': ['pit', 'both', 'pit', 'pit', 'pit', 'pit', 'pit', 'pit', 'pit', 'both', 'both'],
    'model_ID': ['aids3', 'aids5', 'bank3', 'telescope', 'spotify', 'news4', 'slice4', 'energy4', 'housing3', 'housing5', 'housing6']
    }
df = pd.DataFrame(data)
print(df)

df['task_type'] = df['task_type'].astype('category')
df['pit_ordinary'] = df['pit_ordinary'].astype('category') 
df['normal_backtransform'] = df['normal_backtransform'].astype('category') 
df['trainable_bandwidths'] = df['trainable_bandwidths'].astype('category') 

model1 = ols("achieved_error ~ num_classes + C(task_type) + num_features + instances + C(pit_ordinary)", data=df).fit()
anova_table1 = sm.stats.anova_lm(model1, typ=2)
print(anova_table1)

model2 = ols("num_pit_samples ~ C(task_type) + take_num_samples_when_full + C(normal_backtransform) + C(trainable_bandwidths)", data=df).fit()
anova_table2 = sm.stats.anova_lm(model2, typ=2)
print(anova_table2)
```
