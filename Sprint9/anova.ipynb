```python
import pandas as pd

cal_house_model1 = pd.read_csv("sprint5/cal_house_model1.csv")
cal_house_model2 = pd.read_csv("sprint5/cal_house_model2.csv")
cal_house_model3 = pd.read_csv("sprint5/cal_house_model3.csv")

def clean_df(df):
    df.columns = df.iloc[0]
    df = df.drop(df.index[0]) # drop first row
    df_clean = df.drop(df.columns[0], axis=1) # drop first col
    return df_clean

def add_col(df,col_name,value):
    df[col_name] = value
    return df

def col_types(df):
    df["num_pit_samples"] = df["num_pit_samples"].astype(int) 
    df["take_num_samples_when_full"] = df["take_num_samples_when_full"].astype(int) 
    df["normal_backtransform"] = df["normal_backtransform"].astype(bool)
    df["trainable_bandwidths"] = df["trainable_bandwidths"].astype(bool)
    return df


cal_house_model1 = clean_df(cal_house_model1)
cal_house_model2 = clean_df(cal_house_model2)
cal_house_model3 = clean_df(cal_house_model3)

cal_house = pd.concat([cal_house_model1, cal_house_model2, cal_house_model3], ignore_index=True)

cal_house['mae'] = cal_house['mae'].astype(float)
cal_house['r2'] = cal_house['r2'].astype(float)
#cal_house = col_types(cal_house)

cal_house = add_col(cal_house,"task_type","regression")
cal_house = add_col(cal_house,"num_classes",-1)
cal_house = add_col(cal_house,"dataset","cal_house")

min_mae = cal_house["mae"].min()
max_mae = cal_house["mae"].max()
min_r2, max_r2 = cal_house["r2"].min(), cal_house["r2"].max()

cal_house['normalized_mae'] = 1 - (cal_house['mae'] -min_mae) / (max_mae - min_mae)
cal_house['normalized_R2'] = (cal_house['r2'] - min_r2) / (max_r2 - min_r2)

cal_house["error"] = (cal_house["normalized_mae"]+cal_house["normalized_R2"])/2

cal_house = cal_house.drop(["mae","r2","normalized_mae","normalized_R2"],axis=1)
cal_house["number_features"] = 8
cal_house["instances"] = 20640
print(cal_house.dtypes)

cal_house.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

everything_df = pd.read_csv('everything.csv')

everything_clean = everything_df.drop(everything_df.columns[0], axis=1) # drop first col
everything = pd.concat([everything_clean,conductivity],ignore_index=True)

# ANOVA formula
# formula = 'error ~  C(take_num_samples_when_full) + C(trainable_bandwidths)'
formula = 'error ~  C(task_type) + C(model_id) + C(num_classes) + C(dataset) + C(number_features) + C(instances)'
# formula = 'error ~  C(model_id) + C(num_pit_samples) + C(trainable_bandwidths)'
# Fit the model
model = ols(formula, data=everything).fit()


# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```
